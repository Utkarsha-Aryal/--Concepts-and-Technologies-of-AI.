```python
# 3.2 Exercises - Data Cleaning and Transformations with ”Titanic Dataset”:
import pandas as pd
df = pd.read_csv(r"C:\Users\ACER\Titanic-Dataset.csv")
subset_df  = df[['Name','Pclass','Sex','Age','Fare','Survived']]
first_class_passengers = subset_df[subset_df["Pclass"] == 1]

# Calculate statistics for 'Fare'
fare_mean = first_class_passengers["Fare"].mean()
fare_median = first_class_passengers["Fare"].median()
fare_max = first_class_passengers["Fare"].max()
fare_min = first_class_passengers["Fare"].min()
print(f"Mean Fare: {fare_mean}")
print(f"Median Fare: {fare_median}")
print(f"Maximum Fare: {fare_max}")
print(f"Minimum Fare: {fare_min}")
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\ACER\Titanic-Dataset.csv")
subset_df = df[["Name", "Pclass", "Sex", "Age", "Fare", "Survived"]]

# Filter rows where 'Pclass' is equal to 1
first_class_passengers = subset_df[subset_df["Pclass"] == 1]

# Count null values in the 'Age' column
null_count = first_class_passengers["Age"].isnull().sum()
print(f"Number of null values in 'Age' column: {null_count}")

# Drop rows with null values in the 'Age' column
cleaned_df = first_class_passengers.dropna(subset=["Age"])

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(cleaned_df)
```

```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv(r"C:\Users\ACER\Titanic-Dataset.csv")

# One-hot encode the 'Embarked' column
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')

# Add the new columns to the original DataFrame
df = pd.concat([df, embarked_dummies], axis=1)

# Drop the original 'Embarked' column
df.drop('Embarked', axis=1, inplace=True)

# Print the first few rows of the modified DataFrame
print("Modified DataFrame:")
print(df.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = pd.read_csv(r"C:\Users\ACER\Titanic-Dataset.csv")

# Group by 'Sex' and calculate mean survival rate
gender_survival = df.groupby('Sex')['Survived'].mean()

# Print mean survival rates
print("Mean Survival Rates by Gender:")
print(gender_survival)

# Create a bar chart
sns.barplot(x=gender_survival.index, y=gender_survival.values, palette="viridis")
plt.title("Mean Survival Rates by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Group by 'Embarked' and 'Sex', and calculate mean survival rates
embarked_gender_survival = df.groupby(['Embarked', 'Sex'])['Survived'].mean().reset_index()

# Print mean survival rates
print("\nMean Survival Rates by Embarkation Port and Gender:")
print(embarked_gender_survival)

# Create a grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=embarked_gender_survival, x='Embarked', y='Survived', hue='Sex', palette="mako")
plt.title("Survival Rates by Embarkation Port and Gender")
plt.xlabel("Port of Embarkation")
plt.ylabel("Survival Rate")
plt.legend(title="Gender")
plt.show()
```
