```python
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).
```

```python

def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(categorize_age)
df.head()
```

 Calculate the total number of passengers on the Titanic

```python
total_passengers = df['PassengerId'].count()
total_passengers
```

 Count the number of passengers in each age group.

```python
df['age_group'].value_counts()
```

. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
adult_proportion = df['age_group'].value_counts()['adult'] / total_passengers
print(adult_proportion*100)
```

```python
senior_proportion = df['age_group'].value_counts()['senior']/total_passengers
print(senior_proportion*100)
```

```python
child_proportion = df['age_group'].value_counts()['child']/total_passengers
print(child_proportion*100)
```
